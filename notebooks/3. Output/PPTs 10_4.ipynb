```python
data = pd.read_csv('../../output/tableformat2000_2010_2020.csv')
data.head(2)
```

```python
data = data[['index', 'Sumner County, Tennessee', 'Wilson County, Tennessee', 'Portland city, Tennessee', 'Millersville city, Tennessee', 'Gallatin city, Tennessee', 
            'Hendersonville city, Tennessee', 'Lebanon city, Tennessee', 'Mount Juliet city, Tennessee', 'Robertson County, Tennessee']]
data = data.set_index('index').transpose().reset_index()
data = data.rename(columns = {'index':'NAME'})
```

```python
df = data[['NAME', 'GEO_ID', 
           'HHIncome:200K or More 2010', 'HHIncome:200K or More 2020', #change in "high earners"
           'Population 2010', 'Population 2020', #current and change in population
           'Employment:Civilian Unemployment Rate% 2020', 'Employment:Civilian Unemployment Rate% 2010',
           'Minority % 2020', 'Age%:School Age 5 to 18 2020', 'Poverty%: Below Poverty Level 2020', 'Age%:65 and Older 2020', #percent groups current
           'Median Household Income 2010', 'Median Household Income 2020', #current and percent change in median hh income
           'Employment:Unemployed Civilian 2010', 'Employment:Unemployed Civilian 2020', #unemployment change 2010-2020
           'Earnings:100,000 and Over 2010', 'Earnings:100,000 and Over 2020', #high earners (above $200K) change 2010-2020
           'Poverty: Population Below Poverty Level 2010', 'Poverty: Population Below Poverty Level 2020', #percent change population in poverty
           'CB%: Cost Burdened Homeowners 2020', 'CB%:Cost Burdened Renters 2020', #current cost burdned homeowners and renters
           'Ind:Employed Civilians 16+ 2010', 'Ind:Employed Civilians 16+ 2020', #employed civilians place of residence
           'Ind%:Blue Collar 2020', 'Ind%:White Collar 2020' #blue and white collar current
          ]] 
cols = ['Population 2010', 'Population 2020', #current and change in population
        'HHIncome:200K or More 2010', 'HHIncome:200K or More 2020', #percent change in high income households
        'Minority % 2020', 'Age%:School Age 5 to 18 2020', 'Poverty%: Below Poverty Level 2020', 'Age%:65 and Older 2020', #percent groups current
        'Median Household Income 2010', 'Median Household Income 2020', #current and percent change in median hh income
        'Employment:Civilian Unemployment Rate% 2020', 'Employment:Civilian Unemployment Rate% 2010',
        'Employment:Unemployed Civilian 2010', 'Employment:Unemployed Civilian 2020', #unemployment change 2010-2020
        'Earnings:100,000 and Over 2010', 'Earnings:100,000 and Over 2020',#high earners (above $200K) change 2010-2020
        'Poverty: Population Below Poverty Level 2010', 'Poverty: Population Below Poverty Level 2020', #percent change population in poverty
        'CB%: Cost Burdened Homeowners 2020', 'CB%:Cost Burdened Renters 2020', #current cost burdned homeowners and renters
        'Ind:Employed Civilians 16+ 2010', 'Ind:Employed Civilians 16+ 2020', #employed civilians place of residence
        'Ind%:Blue Collar 2020', 'Ind%:White Collar 2020' #blue and white collar current
       ]
df[cols] = df[cols].astype(float)
```

```python
#df.head()
```

```python
#percent change population 2010-2020
df['Percent Increase in Population 2010 to 2020'] = percentchange(df['Population 2020'], df['Population 2010'])
#population growth per month 2010-2020
totalmonths = 120
diffpeople2010to2020 = df['Population 2020'] - df['Population 2010']
df['People Per Month Between 2010 and 2020'] = diffpeople2010to2020/totalmonths
#percent change in median household income 2010-2020
df['Percent Change in Median Household Income 2010 to 2020'] = percentchange(df['Median Household Income 2020'], df['Median Household Income 2010'])
#percent change in unemployment
df['Percent Change in Unemployed Civilians 2010 to 2020'] = percentchange(df['Employment:Unemployed Civilian 2020'], df['Employment:Unemployed Civilian 2010'])
#percent change in high earners
df['Percent Change in High Earners (100,000 and Up) 2010 to 2020'] = percentchange(df['Earnings:100,000 and Over 2020'], df['Earnings:100,000 and Over 2010'])
#percent change in population in poverty
df['Percent Change in Population Living Below Poverty Level 2010 to 2020'] = percentchange(df['Poverty: Population Below Poverty Level 2020'], 
                                                                                           df['Poverty: Population Below Poverty Level 2010'])
#percent change in high income households
df['Percent Change in High Income Households ($200,000 or More) 2010 to 2020'] = percentchange(df['HHIncome:200K or More 2020'], 
                                                                                              df['HHIncome:200K or More 2010'])
#percent change in unemployment rate
df['Percent Change in Unemployment Rate 2010 to 2020'] = percentchange(df['Employment:Civilian Unemployment Rate% 2020'], 
                                                                       df['Employment:Civilian Unemployment Rate% 2010'])


cols = ['Population 2010', 'Ind:Employed Civilians 16+ 2010', 'Median Household Income 2010', 'Poverty: Population Below Poverty Level 2010',
        'HHIncome:200K or More 2010', 'HHIncome:200K or More 2020',
        'Poverty: Population Below Poverty Level 2020', 'Employment:Unemployed Civilian 2020', 'Employment:Unemployed Civilian 2010', 
        'Earnings:100,000 and Over 2020', 'Earnings:100,000 and Over 2010'
       ]
df = df.drop(columns = cols)
```

```python
df = df.rename(columns = {'Population 2020':'Population: Current Population', 'Minority % 2020':'Population: Percent Ethnic Minorities',
                          'Age%:School Age 5 to 18 2020':'Population: Percent School-Aged Children', 
                          'Poverty%: Below Poverty Level 2020':'Population: Percent Living Below Poverty', 
                          'Age%:65 and Older 2020':'Population: Percent Senior Citizens',
                          'Median Household Income 2020':'Housing: Countywide Median Household Income', 
                          'CB%: Cost Burdened Homeowners 2020':'Housing: Percent of Homeowners Who are Costburdened', 
                          'CB%:Cost Burdened Renters 2020':'Housing: Percent of Renters Who are Costburdened', 
                          'Ind:Employed Civilians 16+ 2020':'Employment ACS: Current Number of Occupied Jobs', 
                          'Ind%:Blue Collar 2020':'Employment ACS: Occupied Percent Blue Collar',
                          'Ind%:White Collar 2020':'Employment ACS: Occupied Percent White Collar', 
                         })
```
